# 소셜 네트워크 분석
소셜 네트워크 분석은 사회 구조와 상호 의존성을 파악할 때, 개인, 그룹, 기관 작업, 입력에 대한 출력 패턴 검토에 유용하다. 또한 어떤 사람들이 관계망 가능에 중요한 역할을 담당하는지, 어떤 하위 그룹이 있는지, 특정 관계망에 어떤 연결 관계가 있는지를 분석하여 관계와 개인 사이의 상대적인 위치를 발견할 수 있다.

---
### 네트워크 이론의 개요 및 역사
+ 네트워크: 각각의 객체들이 상호 연결된 구조  
+ 구성요소: node(node, vertex)와 링크(link, edge)  
+ 노드: 고유한 속성을 가지고 있는 행위자  
+ 링크: 노드들간의 관계(relation)  

__-> 네트워크는 복수의 노드들이 연결된 구조__

---
### 네트워크 이론의 역사
+ __오일러의 '쾨니히스베르크 다리 건너기 문제__  
그래프 이론의 발전에 있어 탄탄한 수학적 기초를 제공

+ __모레노의 소시오 그램__  
사람들간의 상호관계를 과학적 분석의 대상으로 발전, 소셜 네트워크 분석 연구의 초석이 됨. 사회 측정법인 소시오메트리로 발전

+ __바베라스와 리빗의 실험__  
중심을 가지고 있는 스타형 과 Y형 구조가 문제를 가장 빨리 해결, 의사전달 과정에서 오류도 가장 적었음

+ __밀그램의 6단계 분리이론__  
좁은 세상 효과, 편지가 목표 대상자에게 평균 5.5단계를 거쳐 6단계 이론이 등장

+ __그라노베터의 '약한 연결의 강점'__  
약한 사회적 관계의 유용성에 대한 새로운 시사점을 제시

+ __프리만의 '중심성 지표' 개발__  
연결정도 중심성, 근접 중심성, 매개 중심성이 네트워크 분석에 있어 가장 널리 상용되는 척도가 됨

+ __버트의 '구조적 공백'__  
구조적 공백의 위치에 있는 노드는 노드와 노드를 연결시켜 주는 역할을 함, 약한 연결 개념을 더 구체화 한 것

+ __왓츠, 스트로가츠의 '좁은 세상 네트워크'__  
정규 네트워크, 무작위 네트워크, 좁은 세상 네트워크를 제시, 노드들 간에 지름길이 존재. 이것을 통해 전체 네트워크가 긴밀하게 연결. 이후 무척도 네트워크가 밝혀지게 되는 계기를 제공

+ __바라바시, 알버트, 정하웅의 '무척도 네트워크'__  
아주 많은 링크를 가진 소수의 노드들이 형성, 이런 노드들은 전체 네트워크에서 허브 등의 중요한 역할을 하게 된다.

# 네트워크 이론의 기초



---
### 그래프의 기본 개념  
+ 그래프 이론: 그래프는 노드(node, vertex)로 구성, 이러한 노드들은 링크(link, edge)로 연결되어 있음  
+ 무방향 네트워크: 책, 논문의 공저자, 네트워크, 페이스북의 친구 관계 네트워크 등  
+ 방향 네트워크: 인터넷 주소의 하이퍼링크 네트워크, 전화통화 네트워크, 트위터의 리트윗, 팔로잉 네트워크 등  

__네트워크 그래프 관계의 정도(강도) 표시 가능__  
+ 이진 네트워크: 관계의 유무만을 표현  
+ 가중 네트워크: 관계의 정도를 함께 표시  


---
### 그래프의 표현 방법  
+ 인접행렬: 인접한 노드들 간의 연결 관계를 행렬로 표현  
+ 노드 리스트: 인접한 링크가 있는 노드만을 표기하는 방법으로 노드들만의 목록  
+ 엣지 리스트: 노드들을 연결한 링크들의 리스트로 '링크의 목록'  

---
# 네트워크 유형

---
+ 네트워크는 링크의 방향성 유무와 가중치 유무에 따라 크게 4가지 유형으로 구분이 가능  
+ 네트워크를 이루고 있는 관계 위상 형태에 따라 구분 가능  
+ 네트워크를 이루고 있는 노드와 링크의 규모 등 네트워크 크기에 따라 구분 가능


+ __분석 대상에 따른 분류__  
전체 네트워크: 네트워크를 구성하고 있는 모든 노드를 포함한 네트워크  
하위 네트워크: 전체 네트워크의 일부분으로 부속되어 있는 것으로 컴포넌트, 파당, 클러스터 등이 해당  
파당: 모든 노드들이 상호간에 직접적으로 연결 되어 있음. 다른 파당에는 포함되어 있지 않은 경우를 말함. 3개 이상의 노드 포함  
클러스터: 유사한 속성을 가지고 있는 노드들의 집합  
에고 네트워크: 분석하고자 하는 특정 노드를 중심에 두고, 그 노드와 직접적으로 연결된 노드들 간의 관계를 네트워크로 표현한 것  

---
# 소셜 네트워크 분석 지표

---
### 네트워크 기본 속성 분석
+ 네트워크의 크기: 구성하는 노드들의 수, 크기가 증가할수록 구조의 복잡성도 증가  
+ 네트워크의 밀도: 구성하고 있는 노드들 간의 연결된 정도, 노드들 간 연결이 많아질수록 밀도가 높아짐  
+ 포괄성: 전체 네트워크의 밀집도를 연결된 노드의 관점에서 표현  
+ 연결정도: 해당 노드에 직접 연결되어 있는 노드들의 개수, 또는 링크의 개수  
+ 연결거리: 노드간 관계(링크)의 강도  
+ 직경: 네트워크 상의 모든 연결거리 중에서 가장 긴 연결거리  
+ 평균연결거리: 네트워크내 모든 노드 쌍의 최단 연결거리를 산출하여 평균값을 계산한 것을 해당 네트워크의 평균 연결거리라고 함  
+ 상호성: 방향 네트워크 상에서 2개의 노드 간의 연결이 서로 쌍방향인 경우(상호적 연결), 단방향인 경우(비상호적 연결)  
+ 군집화 계수: 군집화란 네트워크 상에서 특정 부분의 노드들이 밀접하게 연결되어 다른 부분보다 네트워크 밀도가 높아지는 현상, 군집화 계수는 군집화의 정도를 나타내는 지표  

---
### 중심성 분석
+ 중심성 지표는 소셜 네트워크 분석 지표 중에서 일반적으로 가장 많이 사용되는 지표로 한 행위자가 전체 네트워크에서 중심에 위치하는 정도로 표현하는 지표  
+ 연결정도 중심성: 한 노드가 얼마나 많은 다른 노드들과 연결관계를 맺고 있는 지 측정  
+ 근접 중심성: 한 노드가 얼마나 전체 네트워크의 중심에 근접해 있는지를 측정  
+ 매개 중심성: 한 노드가 다른 노드들 간의 네트워크 관계 형성에 있어서 중개자 혹은 매개자 역할을 얼마나 수행하는 지 측정하는 지표  
+ 아이겐 벡터 중심성: 연결된 다른 노드들의 중요도를 함께 반영한 것

---
# NetworkX 파이썬 패키지를 이용한 네트워크 그래프 작성

```python
import networkx as nx
import matplotlib.pyplot as plt    
%matplotlib inline    
#이미지가 표시되도록 하는 기능
import warnings; warnings.simplefilter('ignore')    #경고 메세지 숨기기
```

+ ### 무방향 네트워크 

```python
G_symmetric = nx.Graph()

G_symmetric.add_edge('Steven',  'Laura')
G_symmetric.add_edge('Steven',  'Marc')
G_symmetric.add_edge('Steven',  'John')
G_symmetric.add_edge('Steven',  'Michelle')
G_symmetric.add_edge('Laura',   'Michelle')
G_symmetric.add_edge('Michelle','Marc')
G_symmetric.add_edge('George',  'John')
G_symmetric.add_edge('George',  'Steven')
```

```python
print(nx.info(G_symmetric))
```

```python
plt.figure(figsize=(5,5))
nx.draw_networkx(G_symmetric);    #네트워크 시각화
```

+ ### 방향 네트워크

```python
G_asymmetric = nx.DiGraph()    #DiGraph를 이용해 방향 네트워크 구축
G_asymmetric.add_edge('A','B')
G_asymmetric.add_edge('A','D')
G_asymmetric.add_edge('C','A')
G_asymmetric.add_edge('D','E')
```

```python
nx.spring_layout(G_asymmetric)    #그래프를 명확히 보기 위해 spring_layout() 함수 사용
nx.draw_networkx(G_asymmetric)
```

+ ### 가중네트워크

```python
G_weighted = nx.Graph()

G_weighted.add_edge('Steven',  'Laura',   weight=25)
G_weighted.add_edge('Steven',  'Marc',    weight=8)
G_weighted.add_edge('Steven',  'John',    weight=11)
G_weighted.add_edge('Steven',  'Michelle',weight=1)
G_weighted.add_edge('Laura',   'Michelle',weight=1)
G_weighted.add_edge('Michelle','Marc',    weight=1)
G_weighted.add_edge('George',  'John',    weight=8)
G_weighted.add_edge('George',  'Steven',  weight=4)
```

```python
elarge =[(u, v) for (u, v, d) in G_weighted.edges(data=True) if d['weight'] > 8]    #가중도 표시 설정
esmall =[(u, v) for (u, v, d)in G_weighted.edges(data=True) if d['weight'] <= 8]

pos = nx.circular_layout(G_weighted)   #노드 전체 

nx.draw_networkx_nodes(G_weighted, pos, node_size=700)    #노드

nx.draw_networkx_edges(G_weighted, pos, edgelist=elarge,width=6)    #링크
nx.draw_networkx_edges(G_weighted, pos, edgelist=esmall,width=6, alpha=0.5, edge_color='b', style='dashed')

nx.draw_networkx_labels(G_weighted, pos, font_size=20, font_family='sans-serif')

plt.axis('off')    #축 제거 
plt.show();
```

+ ### 클러스터링 계수
  
  군집화의 정도를 나타내는 지표를 뜻하며 해당 노드의 이웃 노드들 간에 가능한 최대 관계의 수 중에서 실제로 존재하는 관계의 수가 차지하는 비율로 정의

```python
nx.clustering(G_symmetric,'Michelle')
```

```python
nx.clustering(G_symmetric,'Laura')
```

```python
nx.average_clustering(G_symmetric)    #모든 노드에 대해서 지역 결집계수를 계산하고 그 평균을 구한 것
```

+ ### 네트워크 거리 측정

  1. 연결정도 : 해당 노드에 직접 연결되어 있는 노드들의 개수

```python
nx.degree(G_symmetric, 'Michelle')
```

  2. 연결거리 : 노드간 관계(링크)의 강도

```python
nx.shortest_path(G_symmetric, 'Michelle', 'John')    #두 노드 사이 최단 경로 
```

```python
nx.shortest_path_length(G_symmetric, 'Michelle', 'John')    #두 노드 사이 최단 길이
```

  3. 너비 우선 탐색 : 너비를 우선해서 그래프를 탐색, 정점에서 같은 레벨에 있는 자식 또는 형제들을 탐색하면서 내려가는 알고리즘

```python
S = nx.bfs_tree(G_symmetric, 'Steven')
nx.draw_networkx(S)
```

```python
M = nx.bfs_tree(G_symmetric, 'Michelle')
nx.draw_networkx(M)
```

  4. 이심률 : 노드와 다른 모든 노드 사이의 최대 거리

```python
nx.eccentricity(G_symmetric, 'Michelle')
```

```python
nx.eccentricity(G_symmetric, 'Steven')
```

+ ### 중심성 지표

  1. 연결정도 중심성 : 네트워크 상에서 한 노드가 얼마나 많은 연결 관계를 가지고 있는 지를 측정하는 지표

```python
nx.degree_centrality(G_symmetric)
```

  2. 아이겐벡터 중심성 : 해당 노드에 직접 연결된 다른 노드들의 개수 뿐만아니라 연결된 다른 노드들의 중요도 또한 함께 반영한 것

```python
nx.eigenvector_centrality(G_symmetric)
```

  3. 근접 중심성 : 네트워크 상의 노드들 간의 근접도를 기준으로 정의

```python
nx.closeness_centrality(G_symmetric)
```

  4. 매개 중심성 : 네트워크 상에서 특정 노드가 다른 노드들의 중간에서 얼마나 중개자 및 매개자 역할을 하고 있는 지를 측정하는 지표

```python
nx.betweenness_centrality(G_symmetric)
```

```python
pos = nx.spring_layout(G_symmetric)
betCent = nx.betweenness_centrality(G_symmetric, normalized = True, endpoints = True)
node_color = [20000.0 * G_symmetric.degree(v) for v in G_symmetric]
node_size = [v * 10000 for v in betCent.values()]
plt.figure(figsize=(10,10))
nx.draw_networkx(G_symmetric, pos = pos, with_labels = True, 
                node_color = node_color,
                node_size = node_size)
plt.axis('off');
```

```python
sorted(betCent, key = betCent.get, reverse = True)[:5]
```
